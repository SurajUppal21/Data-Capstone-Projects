### I have also provided pickle file for this data.

```python
allbanks= pd.read_pickle("all_banks")
allbanks
```

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
BAC = data.DataReader("BAC","yahoo",start,end)
BAC.drop("Adj Close",axis=1,inplace=True)
```

```python
C = data.DataReader("C","yahoo",start,end)
C.drop("Adj Close",axis=1,inplace=True)
```

```python
GS = data.DataReader("GS","yahoo",start,end)
GS.drop("Adj Close",axis=1,inplace=True)
```

```python
JPM = data.DataReader("JPM","yahoo",start="Jan 1st 2006",end="Jan 1st 2016")
JPM.drop("Adj Close",axis=1,inplace=True)
```

```python
MS = data.DataReader("MS","yahoo",start,end)
MS.drop("Adj Close",axis=1,inplace=True)
```

```python
WFC = data.DataReader("JPM","yahoo",start,end)
WFC.drop("Adj Close",axis=1,inplace=True)
```

```python
#above we have created 6 dataframes
```

**Creating a list of the ticker symbols (as strings) in alphabetical order.**

```python
tickers = ["BAC","C","GS","JPM","MS","WFC"]
```

**Using pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list.**

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
bank_stocks
```

**Seting the column name levels:**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

**Checking the head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

**What is the max Close price for each bank's stock throughout the time period?**

```python
d = {'num_legs': [4, 4, 2, 2],
     'num_wings': [0, 0, 2, 2],
     'class': ['mammal', 'mammal', 'mammal', 'bird'],
     'animal': ['cat', 'dog', 'bat', 'penguin'],
     'locomotion': ['walks', 'walks', 'flies', 'walks']}
df = pd.DataFrame(data=d)
print("this is before multiindexing \n",df)
df = df.set_index(['class', 'animal', 'locomotion'])
print(df)
df.xs("walks",level=2)#level refers to the index where we want to check
#the class will be level=0 and locomotion will be level=2

#THE ABOVE CAN BE USED TO QUICKLY REVISE XS AND MULTIINDEXING
```

```python
bank_stocks.xs(("BAC"),axis=1)
'''here we could have also passed 
bank_stocks["BAC"] DIRECTLY in the tuple where we have passed in BAC
we can also pass in further info like High,Low,etc'''
bank_stocks["BAC","High"]#this is similar to the XS line above
#we are going to use the provided dataframe to match the results
```

```python
for tick in tickers:
    print(tick,bank_stocks[tick]["Close"].max())
#using XS method below
np.round(bank_stocks.xs(key="Close",axis=1,level=1).max(),2)
#we can also pass in level to be "Stock Info"
#both of the above are the same
```

<b> Creating a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock.
</b>

</b>**We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**</b>

```python
returns=pd.DataFrame()
for tick in tickers:
    returns[tick+" Return"]=allbanks[tick]["Close"].pct_change()
returns.head()
# refer to https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pct_change.html
#to read about the pct_change function
```

**Creating a pairplot using seaborn of the returns dataframe. What stock stands out to you? Can you figure out why?**

```python
sns.set_style("darkgrid")
sns.pairplot(data=returns[1:],diag_kind="hist",diag_kws={"bins":10})
#we can also pass in from where till what we want the data by slicing the data
#kwarg
#as we can see below citigroup is acting weird
#citigroup had a crash in 2008 november
```

**Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. We have noticed that 4 of the banks share the same day for the worst drop, did anything significant happen that day?**

```python
print(returns["BAC Return"].argmin(),"\n")#as we can see argmin only returns the 
# position

# of the index and not the value hence we should use idxmin()
print(returns["BAC Return"].idxmin(),"\n")
# we can also do the same for the whole dataframe
print("min","\n",returns.idxmin(),"\n")
#citigroup around 2011-05-06 announced reverse stock split
```

**We have noticed that Citigroup's largest drop and biggest gain were very close to one another, did anythign significant happen in that time frame**

```python
print("max","\n",returns.idxmax())
#obama was sworn in on 2009-01-20
#stock market can be weird sometimes
```

**Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period? Which would you classify as the riskiest for the year 2015?**

```python
'''sometimes a basic riskiness of a stock can be the measurement of its 
standard deviation'''
print(returns.std())
'''higher std means the stocks are going up and down a lot
#citigroup seems to be the riskiest''';
```

```python
returns.loc["2015-01-01":"2016-01-01"].std()
#ms was the riskiest stock of 2015
```

**Creating a distplot using seaborn of the 2015 returns for Morgan Stanley**

```python
sns.set_style("darkgrid")
sns.distplot(returns.loc["2015-01-01":"2016-01-01"]["MS Return"],color="g",bins=50)
```

**Creating a distplot using seaborn of the 2008 returns for CitiGroup**

```python
sns.distplot(returns.loc["2008-01-01":"2008-12-31"]["MS Return"],color="r",bins=50)
'''the standard deviation is a lot more for this year almost 
8 times of a normal year''';
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

**Ploting the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12,4))
BAC["Close"].loc["2008-01-01":"2009-01-01"].rolling(window=30).mean().plot(label="30 Day moving average")
BAC["Close"].loc["2008-01-01":"2009-01-01"].plot(label="BAC Close")
plt.legend()
#this allows us to see a general trend of the stock as it goes on
```

**Creating a heatmap of the correlation between the stocks Close Price.**

```python
x = bank_stocks.xs(key="Close",level="Stock Info",axis=1).corr()
sns.heatmap(data=x,cmap="coolwarm",annot=True)
```

```python
sns.clustermap(data=x,annot=True,cmap="magma_r")
```

```python
x.iplot(kind="heatmap",colorscale="rdylbu")#refer to docs for more color kwargs
```
